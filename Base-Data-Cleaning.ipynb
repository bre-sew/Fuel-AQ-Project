```python
# dependencies
import os
import pandas as pd
```

```python
# make a list of all files in AirData folder by looping through
AQbyYear = os.listdir("AirData")
df_list = []

for file in AQbyYear:
    df = pd.read_csv(f"AirData/{file}")
    df_list.append(df)
```

```python
# concatenate all csvs in the list
airData = pd.concat(df_list)
airData.reset_index(inplace=True,drop=True)
```

```python
# print all column names
columns=list(airData)
print(columns)
```

```python
# remove extraneous columns
airData = airData[["County Code","Site Num","Parameter Code","Parameter Name","Pollutant Standard","Year",
        "Local Site Name","Address","County Name","City Name"]]
airData = airData.rename(columns={"County Name":"County","City Name":"City",
                                  "Local Site Name":"Site","Parameter Name":"Parameter"})
```

```python
airData
```

```python
sites = airData[["Site Num","Site"]]
sites = sites.drop_duplicates(subset=["Site Num"])
sites
```

```python
# make a list of all files in EMFAC folder by looping through
countyFuel = os.listdir("EMFAC")
df_list = []

for file in countyFuel:
    df = pd.read_csv(f"EMFAC/{file}")
    df_list.append(df)
```

```python
# concatenate all csvs in the list
fuelData = pd.concat(df_list)
fuelData.reset_index(inplace=True,drop=True)
```

```python
# remove extraneous columns
fuelData = fuelData[["Region","Calendar Year","Vehicle Category", "Fuel","Total VMT","Fuel Consumption"]]
fuelData
```

```python
fuelData = fuelData.rename(columns={"Calendar Year":"Year","Region":"County"})
fuelData
```

```python
counties = fuelData["County"].unique()
counties
```

```python
county_df = pd.DataFrame({"County":counties})
county_df
```
