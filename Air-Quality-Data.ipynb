```python
# dependencies
import os
import pandas as pd
```

```python
# make a list of all files in AirData folder by looping through
AQbyYear = os.listdir("AirData")
df_list = []

for file in AQbyYear:
    df = pd.read_csv(f"AirData/{file}")
    df_list.append(df)
```

```python
# concatenate all csvs in the list
airData = pd.concat(df_list)
airData.reset_index(inplace=True,drop=True)
```

```python
airData
```

```python
# print all column names
columns=list(airData)
print(columns)
```

```python
# remove extraneous columns
airData = airData[["County Code","Site Num","Parameter Code","Latitude","Longitude",
        "Parameter Name","Sample Duration","Pollutant Standard","Metric Used","Year",
        "Units of Measure","Arithmetic Mean","1st Max Value","2nd Max Value",
        "Local Site Name","Address","County Name","City Name"]]
airData = airData.rename(columns={"County Name":"County","City Name":"City",
                                  "Local Site Name":"Site","Parameter Name":"Parameter"})
```
